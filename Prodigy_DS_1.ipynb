```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/API_SP_POP_TOTL_DS2_en_csv_v2_3401680.csv")
```

```python
df.head(10)
```

```python
df.tail()
```

```python
df_melted = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                    var_name='Year',
                    value_name='Population')
```

```python
df_melted
```

```python
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
df_melted['Population'] = pd.to_numeric(df_melted['Population'], errors='coerce')
```

```python
df_clean = df_melted.dropna(subset=['Population', 'Year'])
```

```python
df_clean
```

```python
latest_year = df_clean["Year"].max()
df_latest = df_clean[df_clean["Year"]==latest_year]
```

```python
df_latest = df_latest[df_latest["Population"] > 0]
df_latest.head()
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.histplot(df_latest["Population"],bins = 30,kde=True, color='skyblue')

plt.title(f"Distribution of Country Populations in {int(latest_year)}")
plt.xlabel('Population')
plt.ylabel('Number of Countries')

plt.xscale('log')
plt.show()
```

```python
N = 10
df_top_n = df_latest.nlargest(N, 'Population').sort_values('Population')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x="Population",y="Country Name",data = df_top_n,palette="magma")
plt.title(f"Top {N} Most Populous Countries in Year {int(latest_year)}")
# plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
# plt.tight_layout()
plt.show()
```

```python
selected_countries = ['China', 'India', 'USA','Japan','South Korea','Canada']
df_trend = df_clean[df_clean['Country Name'].isin(selected_countries)]

plt.figure(figsize=(12, 6))
for country in selected_countries:
    df_country = df_trend[df_trend['Country Name'] == country]
    plt.plot(df_country['Year'], df_country['Population'], marker='.', label=country)

plt.title('Population Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world.merge(df_latest, left_on='iso_a3', right_on='Country Code')

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column='Population', ax=ax, legend=True,
           cmap='OrRd', linewidth=0.8, edgecolor='0.8')

plt.title(f'Population Heatmap by Country in {int(latest_year)}')
plt.show()
```
